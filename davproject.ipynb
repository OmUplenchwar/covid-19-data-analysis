```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your dataset and it has a publication date column (e.g., 'pubmed_id' or 'doi' might contain dates)

# Convert 'pubmed_id' (or another date column) to datetime format (if applicable)
df['pub_date'] = pd.to_datetime(df['pubmed_id'], errors='coerce')

# Drop missing dates
df_filtered = df.dropna(subset=['pub_date'])

# Group by year and count publications
df_filtered['year'] = df_filtered['pub_date'].dt.year
publication_trend = df_filtered.groupby('year').size().reset_index(name='count')

# Plot line chart
plt.figure(figsize=(10, 5))
sns.lineplot(x=publication_trend['year'], y=publication_trend['count'], marker="o", color="blue")

# Labels and title
plt.xlabel("Year")
plt.ylabel("Number of Publications")
plt.title("Research Publications Over Time")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your dataset

# Create a new column for abstract length
df['abstract_length'] = df['abstract'].astype(str).apply(len)

# Filter for the top 10 most frequent journals
top_journals = df['journal'].value_counts().index[:10]
df_filtered = df[df['journal'].isin(top_journals)]

# Box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x=df_filtered['journal'], y=df_filtered['abstract_length'], palette="Set2")

# Labels and title
plt.xlabel("Journal")
plt.ylabel("Abstract Length (characters)")
plt.title("Abstract Length Distribution Across Journals")
plt.xticks(rotation=45)  # Rotate journal names for better visibility

plt.show()
```
