```python
df = pd.read_csv("metadata.csv", low_memory=False)
print(df.head())
```

```python
print(df.columns)
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')
df.groupby('publish_time')['cord_uid'].count().plot(figsize=(12, 6), title="COVID-19 Research Publications Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Publications")
plt.show()
```

```python
df['journal'].value_counts().head(10).plot(kind='barh', figsize=(12, 6), title="Top 10 Journals Publishing COVID-19 Research")
plt.xlabel("Number of Publications")
plt.ylabel("Journal Name")
plt.gca().invert_yaxis()  # Invert for better readability
plt.show()
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')  # Convert to datetime
df.set_index('publish_time', inplace=True)  # Set date as index
df.resample('M').size().plot(figsize=(12, 6), title="Research Publications Over Time")
plt.xlabel("Publication Date")
plt.ylabel("Number of Papers")
plt.show()
```

```python
top_journals = df['journal'].value_counts().head(10)  # Get top 10 journals
plt.figure(figsize=(12, 6))
sns.barplot(x=top_journals.values, y=top_journals.index, palette="viridis")
plt.xlabel("Number of Papers")
plt.ylabel("Journal Name")
plt.title("Top 10 Journals by Publication Count")
plt.show()
```

```python
top_authors = df['authors'].str.split(';').explode().value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_authors.values, y=top_authors.index, palette="mako")
plt.xlabel("Number of Papers")
plt.ylabel("Author Name")
plt.title("Top 10 Authors by Publication Count")
plt.show()
```

```python
# Split authors by semicolon and count occurrences
from collections import Counter

# Drop missing values in the authors column
df = df.dropna(subset=['authors'])

# Split authors and count occurrences
authors_list = df['authors'].str.split(';').explode()
top_authors = authors_list.value_counts().head(10)

# Plot the top authors
plt.figure(figsize=(12, 6))
sns.barplot(x=top_authors.values, y=top_authors.index, palette='viridis')
plt.xlabel("Number of Papers")
plt.ylabel("Authors")
plt.title("Top 10 Authors with Most Publications")
plt.show()
```
